```python
import numpy as np

# methods fit, 
% matplotlib inline
import matplotlib.pyplot as plt
```

```python
class EatingRabbit:
    def __init__(self, limit=10):
        self.limit = limit #у каждого объекта ф-и свой лимит. это конструктор
        self.counter = 0
    def feed(self, num_carrots):
        self.counter += num_carrots
        if self.counter > self.limit:
            self.counter -= num.carrots
            print('Sorry, I am full')
        else:
            print('Thanks, I have eaten {0} carrots'.format(self.counter))
```

```python
rabbit_1 = EatingRabbit(11)
```

```python
type(rabbit_1)
```

```python
rabbit_1.feed(num_carrots=5)
```

# Обучаем первый классификатор

```python
#artificial vyborka
x_1 = np.random.normal(loc=5.0, scale=1.0, size=(10, 2)) #normalnoe raspredelenie
x_2 = np.random.normal(loc=3.0, scale=1.0, size=(10,2))
y1 = np.ones((10))
y2 = -1 * np.ones((10))

x = np.vstack((x_1, x_2))
y = np.hstack((y1, y2))
```

```python
plt.scatter(x[:,0], x[:,1], c=y, s=100, alpha=0.7)#c - color c=y - столько цветов, сколько игриков. с='g' - все зеленое.
#s - radius. alpha - прозрачность
```

```python
import sklearn
```

```python
#from sklearn.model_selection import train_test_split #newer version
from sklearn.cross_validation import train_test_split #older version
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
#randomstate - псевдослучайные числа, всегда выдает одну и ту же случайную выборку
```

```python
plt.scatter(x_train[:,0], x_train[:,1], c=y_train, s=100, alpha=0.7)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
#need to count accuracy
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
np.mean(y_pred == y_test)
```
